```python
numbers= open('numbers.txt','r')
a=[]
for lines in numbers:
    c=lines.replace('\n','')
    c=c.replace(',',' ')
    for n in c.split():
        a.append(int(n))
freq={}
for item in a:
    if(item in freq):
        freq[item]+=1
    else:
        freq[item]= 1
for key, value in freq.items():
    print("% d :% d"%(key, value))
import matplotlib.pyplot as plt
x=a
plt.hist(x)
plt.show()
import json
y= json.dumps(freq)
with open("midterm.json","w")as outfile:
    json.dump(freq, outfile)



```

```python
import pandas as pd
df =pd.read_csv('amazon.csv')
df=df.fillna(0)
df["Item Total"]=df["Item Total"].str.replace('$','').astype(float)
print("My monthly average for this year was",df["Item Total"].sum()/11)
print("My spend average was", df["Item Total"].mean())
print("My median spending was", df["Item Total"].median())
print("My max money spent on an item was", df["Item Total"].max())
print("My min money spent on an item was",df["Item Total"].min())
print("The standard deviation of my spending was", df["Item Total"].std())
```
